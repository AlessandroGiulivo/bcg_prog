# Data Visualization

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 1, 6, 1]

plt.plot(x, y)

plt.show()
```

```python
import math

axis_font = {'fontname': 'Arial', 'size': '20'}

xs = [0.1*n for n in range(100)]
sinx = [math.sin(x) for x in xs]
cosx = [math.cos(x) for x in xs]

plt.plot(xs, sinx, '.', label = 'sin(x)')
plt.plot(xs, cosx, '--', label = 'cos(x)')
plt.xlabel('x', **axis_font)
plt.ylabel('y', **axis_font)
plt.title('The sin and cos functions')
plt.show()
```

```python
from sklearn import datasets
iris = datasets.load_iris()
x = list(iris.data[:, 2])
y = list(iris.data[:, 3])
plt.scatter(x, y)
plt.show()
```

```python
plt.boxplot([[1,2,3,4], [4, 5, 6, 7, 8]])
plt.yticks([2, 7, 12])
plt.show()
```
