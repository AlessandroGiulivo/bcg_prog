# Classes

```python
import matplotlib.pyplot as plt
```

```python
class Point3D:
    def __init__(self, newx, newy, newz):
        self.x = newx
        self.y = newy
        self.z = newz

class ListPoint3D:
    def __init__(self):
        self.list_of_points = []
        self.length = 0
        
    def append_point(self, point):
        self.list_of_points.append(point)
        if len(self.list_of_points) > 1:
            previous = self.list_of_points[-2]
            delta_x = point.x - previous.x
            delta_y = point.y - previous.y
            delta_z = point.z - previous.z
            delta_dist = (delta_x**2 + delta_y**2 + delta_z**2)**0.5
            self.length = delta_dist
    def plot_line(self):
        ax = plt.axes(projection='3d')
        ax.plot3D([p.x for p in self.list_of_points],
                  [p.y for p in self.list_of_points],
                  [p.z for p in self.list_of_points])
        
```

```python
p1 = Point3D(50, 0, 0)
p2 = Point3D(0, 100, 50)
p3 = Point3D(0, 50, 150)
line = ListPoint3D()

line.append_point(p1)
line.append_point(p2)
line.append_point(p3)

line.plot_line()
```
